### Pandas now - loc, iloc

```python
import pandas as pd
```

```python
data_df = pd.read_csv('data/covid-19-us-counties.csv')
data_df.head()
```

```python
# loc method [rows, columns]. Rows are specified as names of indecies.
data_df.loc[:, :]
```

```python
data_df.loc[:, ['date', 'state']]
```

```python
data_df.loc[[1, 2, 3, 4], ['date', 'state']]
```

```python
# ranges are well supported
data_df.loc[[1, 2, 3, 4], 'date': 'cases']
```

```python
data_df.loc[10:15, 'date': 'cases']
```

```python
# using indicies to display odd and even rows
data_df.loc[data_df.index % 2 == 0, :]
```

```python
# the same principle applies for columns
[i for i in data_df.columns]
```

```python
data_df.loc[data_df.index % 2 == 0, 
            [i for i in data_df.columns if i.startswith('c')]]
```

```python
# iloc() works in the same way, but requires positions instead of names for rows and columns
data_df.iloc[:7, 0:2]
```

```python
# using Category type to save df's memory space, making its size smaller
```

```python
data_df.info(memory_usage='deep')
```

```python
data_df.memory_usage(deep=True)
```

```python
data_df['state'].unique().size
```

```python
# currently we have text field 'state' propagated over 845877 rows. We can convert this field to Category type
# to save some space. We will assign a number to each field, and will use this number in DF
```

```python
data_df['state'] = data_df['state'].astype('category')
```

```python
data_df['state'].head()
```

```python
data_df['state'].cat.codes.head()
```

```python
# checking how much space we saved: went down from 55141566 to 851611!
data_df.memory_usage(deep=True)
```

```python
# we save 98% of initial space!
1 - (851611 / 55141566)
```

```python
# we can also use Category type for sorting. Instead of text fields describing network quality we'll assign a weight to 
# each text field, allowing adequate sorting. So we will have a logic within text fields.
```

```python
# initial df with text field describing network quality
df = pd.DataFrame({'user_id': [1, 2, 3, 4, 5, 6, 7], 
                   'network_quality': ['good', 'very good', 'poor', 'excellent', 'bad', 'good', 'poor']})
df
```

```python
# sorting works but in a weird way - by letters
df.sort_values('network_quality')
```

```python
# creating a custom type for that
nq_type = pd.api.types.CategoricalDtype(categories=['bad', 'poor', 'good', 'very good','excellent'], ordered=True)
```

```python
# converting our df's field into new custom category type
df['network_quality'] = df['network_quality'].astype(nq_type)
```

```python
df
```

```python
# we see the logical link between category levels
df['network_quality'].head(10)
```

```python
# now sorting works properly
df.sort_values('network_quality')
```

```python
# and we can use math operations with the category field now
df.loc[df['network_quality']>= 'good', :]
```
