### Exploring YouTube videos's data set:
#### - calculate unique video categories
#### - calculate mean view for each category
#### - define a separate dataset for top-liked video for each category, save it to a .csv

```python
import pandas as pd
import numpy as np
```

```python
yt_df = pd.read_csv('data/youtube_videos.csv')
```

```python
# exploring data
yt_df.info()
```

```python
yt_df.head(10)
```

```python
# calculate unique video categories
yt_df['category_id'].unique().size
```

```python
yt_df['category_id'].unique()
```

```python
# null values are present in the data set
yt_df[yt_df['category_id'].isnull()]
```

```python
# cleaning null rows and saving into a separate dataset, just in case
yt_df_no_null_rows = yt_df.drop(index=yt_df[yt_df['category_id'].isnull()].index, axis=False)
yt_df_no_null_rows.isnull().value_counts(normalize=True)
```

```python
# calculate mean view for each category
for cat in yt_df['category_id'].unique():
    mean_views = yt_df[yt_df['category_id'] == cat]['views'].mean()
    print(f'Category {cat}, mean view: {round(mean_views, 2):,}')
```

```python
# using dict to display items sorted
views_per_cat = {}
for cat in yt_df['category_id'].unique():
    mean_views = yt_df[yt_df['category_id'] == cat]['views'].mean()
    views_per_cat[cat] = round(mean_views, 2)

views_per_cat_sorted = sorted(views_per_cat.items(), key=lambda x: x[1], reverse=True)

for item in views_per_cat_sorted:
    cat, views = item[0], item[1]
    if not(np.isnan(cat)):
        print(f'Category {cat}, mean view: {views:,}')
```

```python
# define a separate dataset for top-liked video for each category, save it to .csv
```

```python
most_liked_videos = []
for cat in yt_df['category_id'].unique():
    max_likes = yt_df[yt_df['category_id'] == cat]['likes'].max()
    print(f'Category {cat}, max likes: {round(max_likes, 2):,}')
    most_liked_videos.append((yt_df[(yt_df['category_id'] == cat) & (yt_df['likes'] == max_likes)]))
```

```python
len(most_liked_videos)
```

```python
# uniting all separate df's from the list into a single one
top_liked_videos_df=pd.concat(most_liked_videos)
```

```python
top_liked_videos_df
```

```python
# simulating saving to .csv file using sys module
import sys
top_liked_videos_df.to_csv(sys.stdout, index=False)
```
