```python
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_csv('data/covid-19-us-counties.csv')
```

```python
data_df.head()
```

```python
data_df.groupby(['state'])['cases'].sum()
```

```python
# .groupby() returns iterator with tuples, [0] contains the item we grouped by, [1] - DF for that item.
# e.g. for our grouping by states returns, [0] - state, [1] - DF for that state
for item in data_df.groupby(['state']):
    print(f'{item[0][0]}: {item[1]["cases"].sum():,}')
```

```python
# grouping by 2+ fields. We'll have a unique comnination of 2+ fields in [0]
for item in data_df.groupby(['state', 'county']):
    print(f'{item[0]}: {item[1]["cases"].sum():,}')
```

```python
# no need to use 'for' cycles, just indicate the necessary agregation field after:
data_df.groupby(['state', 'county'])['cases'].sum()
```

```python
# executing multiple math methods in agregations with .agg() method
data_df.groupby(['state', 'county']).agg({'cases': ['sum', 'max', 'min', 'mean']})
```

```python
# as .agg() returns a DF, not just Series, we can add more fields for agregation:
data_df.groupby(['state', 'county']).agg({'cases': ['sum', 'max', 'min', 'mean'],
                                          'fips': ['sum']})
```

```python
# index will be a combination of state and county:
data_agg = data_df.groupby(['state', 'county']).agg({'cases': ['sum', 'max', 'min', 'mean'],
                                          'fips': ['sum']})
data_agg.index
```

```python
# we can reset index to have a default digit-based index, basically moving our composite index into DF:
data_agg.reset_index()
```

```python
# we can create our own agregation funcs, using anonymous lambda (can be also used for formatting):
data_df.groupby(['state', 'county']).agg({'cases': ['mean', lambda values: (values>0).sum()]})
```

```python
# (can be also used for formatting)
data_df.groupby(['state', 'county']).agg({'cases': ['mean', lambda values: '$ '+ str((values>0).sum())]})
```

#### - stack & unstack

```python
data_agg = data_df.groupby(['state', 'county']).agg({'cases': ['sum', 'max', 'min', 'mean'],
                                          'fips': ['sum']})
data_agg
```

```python
# last index get converted into column
data_agg.unstack(fill_value=0)
```

```python
# moving columns into index:
data_agg.stack()
```

#### - rolling, shift, sample

```python
weather = pd.DataFrame({'date': ['5/1/2017', '5/2/2017', '5/3/2017', '5/4/2017', '5/5/2017', '5/6/2017', '5/7/2017', '5/8/2017', '5/9/2017'],                        
                        'temperature': [34, 54, 43, 54, 30, 54, 52, 46, 40],
                        'cloudy': [1, 2, 3, 4, 5, 6, 7, 8, 9],
                        'snowing': ['Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No', 'No', 'No', 'No']})
weather
```

```python
# using .rolling() to calculate mean for some window timeframe, e.g. for temperature
weather['temperature'].rolling(window=2).mean()
```

```python
# shifting some data for X steps
weather['temperature'].shift(1)
```

```python
# .shift() is ususally used when we want to compare data YoY, WoW, etc. We can add an additional column, so calc is easier
weather['prev_temperature'] = weather['temperature'].shift(1)
weather
```

```python
illinois_df = data_df[data_df['state'] == 'Illinois'].copy()
illinois_df
```

```python
# using .sample() to limit DF's data to X %%, to make calculations easier & faster
# it returns a random rows from source DF
illinois_df.sample(5)
```

```python
# random_state=XX param allows getting the same random rows all the time. It's like a stable random sample of data,
# to ensure data interpretability while running optimization tasks on different devices
illinois_df.sample(5, random_state=55)
```

```python
# use frac=.xx to indicate required sampling percentage
illinois_df.sample(frac=0.1, random_state=55)
```

#### - concatination of DataSets (concat, join, merge, etc)


```python
illinois_df = data_df[data_df['state'] == 'Illinois'].copy()
washington_df = data_df[data_df['state'] == 'Washington'].copy()
```

```python
# appending one DF to the end of another
pd.concat([washington_df, illinois_df])
```

```python
# or we can specify the axis for concat with axis
pd.concat([washington_df, illinois_df], axis=1)
```

```python
# but if we reset indices for both source DFs, we'll have the following: colums of the second DF will be added to the right 
washington_df.reset_index(inplace=True)
illinois_df.reset_index(inplace=True)
pd.concat([washington_df, illinois_df], axis=1)
```

```python
left = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7),
                    'other_data': range(7)})
right = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3),
                    'other_data': range(3)})
```

```python
left
```

```python
right
```

```python
pd.concat([right, left], axis=1, keys=['right', 'left'])
```

```python
pd.concat([left, right], keys=['left', 'right'], axis=1)
```

```python
# using .join() method, a SQL analogy
right.set_index('key', inplace=True)
```

```python
right
```

```python
# use lsuffix & rsuffix to rename fields with the same name
left.join(right, on='key', lsuffix='_left', rsuffix='_right')
```

```python
# we can define how to join, e.g. 'inner' join - we'll get only those lines where there's a match
left.join(right, on='key', lsuffix='_left', rsuffix='_right', how='inner')
```

```python
# merge allows joining DF's by different key fields
left = pd.DataFrame({'key_1': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7),
                    'other_data': range(7)})
right = pd.DataFrame({'key_2': ['a', 'b', 'd'],
                    'data2': range(3),
                    'other_data': range(3)})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, left_on='key_1', right_on='key_2')
```

```python
# difference bw .join() & .merge()
# .join() - we're joining left DF with right DF by index
# .merge() - allows to specify fields we're using for merge operation. Thus it's a more flexible .join()
```

```python
pd.merge(left, right, left_on='key_1', right_on='key_2', how='outer', suffixes=("_l", "_r"))
```

#### - .apply()


```python
# we use .apply() to run a specific/custom func by rows (axis=1) or by columns (axis=0)
# data_df.apply(function, axis=1)
```

```python
data_df.head()
```

```python
def row_to_str(row):
    return row['county'] + ' - ' + row['state']
```

```python
data_df.apply(row_to_str, axis=1).head()
```

```python
# doing the same by using lambda func
data_df.apply(lambda row: row['county'] + ' - ' + row['state'], axis=1).head()
```

```python
# applying by column
data_df['cases'].apply(lambda x: x>10)
```

```python
data_df['cases > 10'] = data_df['cases'].apply(lambda x: x>10)
```

```python
data_df
```

```python
# we can agregate by columns, calculating sum or mean, and keep original amount of rows with result_type='broadcast'
data_df[['cases', 'fips']].apply(lambda x: x.mean())
```

```python
data_df[['cases', 'fips']].apply(lambda x: x.mean(), result_type='broadcast')
```

####  - pivot vs pivot_table

```python
weather = pd.DataFrame({'date': ['5/1/2017', '5/2/2017', '5/3/2017', '5/1/2017', '5/2/2017', '5/3/2017', '5/1/2017', '5/2/2017', '5/3/2017'],
                        'city': ['New York', 'New York', 'New York', 'London', 'London', 'London', 'Tokio', 'Tokio', 'Tokio'],
                        'temperature': [34, 54, 43, 54, 30, 54, 52, 46, 40],
                        'cloudy': [1, 2, 3, 4, 5, 6, 7, 8, 9],
                        'snowing': ['Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No', 'No', 'No', 'No']})
```

```python
weather
```

```python
# we structure/reshape source data as we want by using .pivot(). Data agregation is NOT supported - use pivot_table()
weather.pivot(index='date', columns='city', values='temperature')
```

```python
# if we need to agregate, we have to use .pivot_table
weather.pivot_table(index='snowing', columns='city', values='temperature', aggfunc='sum', fill_value=0)
```
