```python
# creating DataFrame from dict
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
```

```python
pd.DataFrame(data)
```

```python
# Specifying columns' order explicitly
pd.DataFrame(data=data, columns=['year', 'state', 'pop'])
```

```python
# we can specify unique text index too - number of idx values should be equal to number of rows
pd.DataFrame(data=data, columns=['year', 'state', 'pop'],
             index=['one', 'two', 'three', 'four', 'five', 'six'])
```

```python
# we can add columns too, NaN will be values
frame = pd.DataFrame(data=data, columns=['year', 'state', 'pop', 'debt', 'col3'],
             index=['one', 'two', 'three', 'four', 'five', 'six'])
```

```python
frame
```

```python
frame['debt'] = 16.5
```

```python
frame
```

```python
# calculating value of a columns based on other columns' values
frame['col3'] = frame['pop'] + frame['debt']
```

```python
frame
```

```python
# creating a new column out of numpy array
frame['col_4'] = np.arange(6)
```

```python
frame
```

```python
frame.index
```

```python
frame.values
```

```python
frame.values.shape
```

```python
# creating a new dataframe from an existing one with copy()
temp_frame = frame.copy()
```

```python
# deleting a column with a system method del
del temp_frame['col_4']
```

```python
temp_frame
```

```python
frame
```

```python
# use drop() method for deleting multiple columns. axis tells where rows or columns (axis=True) will be affected
temp_frame.drop(['col3', 'debt'], axis=True)
```

```python
# deleted columns still exist as we have not assigned new values to a dataframe - use inplace=True to keep changes
temp_frame
```

```python
# using inplace to keep changes
temp_frame.drop(['col3', 'debt'], axis=True, inplace=True)
```

```python
temp_frame
```

```python
# deleting rows, axis=False
temp_frame.drop(['one', 'two', 'four'], axis=False)
```

```python
# deleting dataframe as a whole
del temp_frame
```

```python
frame
```

```python
# filtering rows with True/False list - e.g. showing only those rows we need
[True, False, True, True, False, False]
```

```python
frame[[True, False, True, True, False, False]]
```

```python
# or we can create our own True/False filter based on condition:
frame['pop'] >= 2.9
```

```python
# and use this custom filter:
frame[frame['pop'] >= 2.9]
```

```python
# additional filtering - display only columns we need:
frame[frame['pop'] >= 2.9][['year', 'state']]
```

```python
# filtering using multiple conditions - frame[(cond_1) & (cond_2)]
# first condition:
frame['state'] == 'Ohio'
```

```python
frame[(frame['state'] == 'Ohio') & (frame['pop'] >= 2.9)]
```

```python
# using .isin() method
frame['year'].isin([2000, 2003])
```

```python
frame[frame['year'].isin([2000, 2003])]
```
