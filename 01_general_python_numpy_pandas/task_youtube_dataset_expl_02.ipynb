### Exploring YouTube videos's data set:
##### - calculate mean\min\max views for each category using groupby
##### - calculate tags for each video
##### - add an additional field 'title_category' using descripton stored in **category_id.csv**
##### - use **pivot_table** to calculate views for some video channels by category titles.

```python
import pandas as pd
```

```python
yt_df = pd.read_csv('data/youtube_videos.csv')
```

```python
yt_df.head()
```

```python
# - calculate mean\min\max views for each category using groupby
yt_df.groupby('category_id',sort=False).agg({'views':['mean', 'min','max']})
```

```python
# - calculate tags for each video
yt_df.groupby('video_id')['tags'].apply('|'.join).apply(lambda row: len(set(str(row).split("|"))))
```

```python
# - add an additional field 'title_category' using descripton stored in **category_id.csv**
categories_df = pd.read_csv('data/category_id.csv')
```

```python
yt_df = pd.merge(yt_df, categories_df, on='category_id', suffixes=(None,'_category'))
```

```python
yt_df.head()
```

```python
# - use **pivot_table** to calculate views for some video channels by category titles.
yt_df[(yt_df['channel_title']=='Bruno Mars')|
      (yt_df['channel_title']=='BBC News')|
      (yt_df['channel_title']=='Jimmy Kimmel Live')].pivot_table(index='channel_title', 
                                                                 columns='title_category',
                                                                 values='views', 
                                                                 aggfunc='sum',
                                                                 fill_value=0,
                                                                 dropna=True)
```
